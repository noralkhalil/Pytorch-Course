```python
import torch
# import numpy as np
```

```python
tensorA = torch.tensor([1,2,3,4])
tensorB = torch.tensor([5,6,7,8])
```

```python
# addition:
tensorA + tensorB
```

```python
# addition Longhad:
torch.add(tensorA, tensorB)
```

```python
# Subtraction:
tensorA - tensorB
```

```python
# Sub Function:
torch.sub(tensorA, tensorB)
```

```python
# Multiplication:
tensorA * tensorB
```

```python
# Mul longhad:
torch.mul(tensorA, tensorB)
```

```python
# Devision:
tensorA / tensorB
```

```python
# Div Longhad:
torch.div(tensorA, tensorB)
```

```python
# Remainder Moduls:
tensorB % tensorA
```

```python
# Rmeainder Longhad:
torch.remainder(tensorB, tensorA)
```

```python
# Exponents / Power
torch.pow(tensorA, tensorB)
```

```python
# Another way to wirte longhad
tensorA.add(tensorB)
```

```python
# Reassignment:
tensorA + tensorB
```
