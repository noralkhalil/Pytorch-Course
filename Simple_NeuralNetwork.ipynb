```python
import torch
import torch.nn as nn
import torch.nn.functional as F # it helpa us to bring the date forward, from an hidden layer to the other hidden layer
```

```python
# Create a Model Class that inherits nn.Module:
class Modelll(nn.Module):

  # --> Input Layer (4 features of the flower)

  # --> Hidden_Layer_1 (Number of neuerons)
  # --> Hidden_Layer_2 (Number of neuerons)
  # --> Hidden_Layer_n (Number of neuerons)

  # --> Output (3 classes of Iris flowers)

  def __init__(self, in_features = 4 , h1=8, h2=9, out_features=3):

        # h1 = Anzahl der Neuronen im ersten Hidden Layer
        # h2 = Anzahl der Neuronen im zweiten Hidden Layer

    super().__init__()    # to instantiate our nn.Module

    self.fc1 = nn.Linear(in_features, h1)     # fc1: Es ist "Fully Connected Layer 1" , Es gibt die infos von Eingang (in_feature) zu Hidden_Layer_1
    self.fc2 = nn.Linear(h1, h2)              # fc2: Es ist "Fully Connected Layer 2" , Es gibt die infos von Hidden_Layer_1 zu Hidden_Layer_2
    self.out = nn.Linear(h2, out_features)    # out: Es ist Ausgang , Es gibt die infos von Hidden_Layer_2 zu Ausgang (out_feature)

                                              # Linear: == Matrix × Vektor + Bias

    # Now We need a function that moves everything forwared:
  def forward(self, x):
    x = F.relu(self.fc1(x))          # relu: ReLU, das ist ein activation function
    x = F.relu(self.fc2(x))
    x = self.out(x)

    return x


```

```python
# Pick a manaul seed for randomization:       // damit wir ungefähr die gleiche randum_Werte bekommen.
torch.manual_seed(41)

# Create an instance of model:                // d.h. eine kopie von unsere Class machen
New_Copy_model = Modelll()
```

```python
# Train Test split! Set X,Y !!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!

X = myDataFram.drop('species', axis=1)
Y = myDataFram['species']
```

```python
# Convert these to numpy arrays: !!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
X = X.values
Y = Y.values
```

```python
# Standard Zeile: !!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
from sklearn.model_selection import train_test_split
```

```python
# Train test split
```

```python
# Standard Zeile:!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
XTrain, XTest, YTrain, YTest = train_test_split(X, Y, test_size=0.2, random_state=41)
```

```python
# Convert X and Y-features to float Tensors: !!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
XTrain = torch.FloatTensor(XTrain)
XTest = torch.FloatTensor(XTest)
YTrain = torch.LongTensor(YTrain)
YTest = torch.LongTensor(YTest)
```

```python
# Set the criterion of model to measure the error, how far off the predicions are from the data
criterion = nn.CrossEntropyLoss()                   #!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!

# Choose Optimizer (we use here Adam optimizer),
# Choose Learning rate = lr: --> d.h. if error does not go down after a bunch of iterations (epochs), lower our learning rate

optimizer = torch.optim.Adam(New_Copy_model.parameters(), lr=0.01)
```

```python
# Train our model:
# 1. How many Epoches do we wnat? (Epoche: is one run thru all the training data in our network)
epochs = 100
losses = []
for i in range(epochs):
  # Go forward and get a prediction
  i = i + 1
  y_pred = New_Copy_model.forward(XTrain)
  # Measure the loss/error, gona be high at first
  loss = criterion(y_pred, YTrain)    # predicted values vs YTrain
  # Keep Track of our lossees:
  losses.append(loss.detach().numpy)

  # Print every 10 epochs:
  if i % 10 == 0:
    print(f'Epoch number: {i} and the loss: {loss}')

 # Do some back propagation: take the error rate forward propagation and feed it back
 # thru the network to fine tune the weifhts
  optimizer.zero_grad()
  loss.backward()
  optimizer.step()
```
